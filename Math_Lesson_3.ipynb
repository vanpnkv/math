```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)
X, Y = np.meshgrid(X, Y)
Z = -10*X + 10*Y
Z1 = -10*X + 10*Y + 200
ax.plot_wireframe(X, Y, Z)
ax.plot_wireframe(X, Y, Z1)
show()
```

```python
# Эллиптический параболоид
fig = figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)
X, Y = np.meshgrid(X, Y)
Z = (X**2 + Y**2) / 2
ax.plot_surface(X, Y, Z)
ax.plot_wireframe(X, Y, Z)
show()
```

```python
# Гиперболический параболоид
fig = figure()
ax = Axes3D(fig)
X = np.arange(-10, 10, 0.5)
Y = np.arange(-10, 10, 0.5)
X, Y = np.meshgrid(X, Y)
Z = (X**2 - Y**2) / 2
ax.plot_surface(X, Y, Z)
ax.plot_wireframe(X, Y, Z)
show()
```

### 4. Задание (в программе)
Решите систему уравнений:  
y = x^2 – 1  
exp(x) + x∙(1 – y) = 1  


Решите систему уравнений и неравенств:  
y = x^2 – 1  
exp(x) + x∙(1 – y) > 1  


```python
x = np.linspace(-10, 10, 100)
plt.plot(x, x**2 - 1)
plt.plot(x, (exp(x) + x - 1) / x)
plt.xlim(-5, 5)
plt.ylim(-2, 10)
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import fsolve
def f(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)
x1, y1 = fsolve(f, (-2, 2))
print(x1, y1)
x1, y1 = fsolve(f, (2, 4))
print(x1, y1)
```

```python
x1, y1 = fsolve(f, (1, 4))
print(x1, y1)
```
